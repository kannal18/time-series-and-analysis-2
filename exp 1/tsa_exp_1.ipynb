## Data wrangling

### Subtask:
Convert the 'DATE_TIME' column to a datetime object and set it as the index of the DataFrame.


**Reasoning**:
Convert the 'DATE_TIME' column to a datetime object and set it as the index of the DataFrame.



```python
# Convert 'DATE_TIME' column to datetime objects (if not already done)
if not pd.api.types.is_datetime64_any_dtype(df['DATE_TIME']):
  df['DATE_TIME'] = pd.to_datetime(df['DATE_TIME'])

# Set 'DATE_TIME' as the index
df.set_index('DATE_TIME', inplace=True)

# Sort the index in ascending order (optional)
df.sort_index(inplace=True)

display(df.head())
```
